```python
import os
import zipfile
from ultralytics import YOLO

from google.colab import drive
drive.mount('/content/drive')
```

```python
zip_path = "/content/drive/MyDrive/car-plate-detector-data.zip"

extract_path = "/content/dataset"

os.makedirs(extract_path, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
```

```python
yaml_content = """
train: /content/dataset/car-plate-detector-data/obj_Train_data/train
val: /content/dataset/car-plate-detector-data/obj_Validation_data/val
test: /content/dataset/car-plate-detector-data/obj_Test_data/test

nc: 1
names: ["plate"]
"""

with open('/content/data.yaml', 'w') as f:
    f.write(yaml_content)

```

```python
model = YOLO('yolov8s.pt')

model.train(
    data='/content/data.yaml',
    epochs=30,
    batch=32,
    imgsz=640,
    save_period=10,
    save=True,
    device=0
)
```
